## Use groupby() to get the type of sales correspond to the customer name with the values

```python
x = df_lunas_no_dupli.groupby('MATERIAL / JASA')['CUSTOMER'].value_counts().to_frame()
x.rename(columns={'CUSTOMER': 'value'}, inplace = True)
x.reset_index(inplace=True) 
x
```

## Create a dataframe contains all the sales column + the Total Sales from previous

```python
customer_sort_index = customer_value.sort_index()
customer_sort_index.head()
```

```python
x2 = x.pivot_table(index =['CUSTOMER'], columns = 'MATERIAL / JASA'
                   , values = 'value').reset_index()
x2.fillna(0, inplace = True)
x2.iloc[:,1:] = x2.iloc[:,1:].astype('int32')

x2['MATERIAL']= x2['MATERIAL'] + x2['MATERIAL ']
x2.drop(columns = 'MATERIAL ', inplace = True)

x2['MATERIAL & JASA']= x2['MATERIAL & JASA'] + x2['JASA & MATERIAL']
x2.drop(columns = 'JASA & MATERIAL', inplace = True)

x2.set_index('CUSTOMER', drop = True, inplace = True)
x2

## Swap column Material <-> Jasa

cols = list(x2.columns)
a, b = cols.index('JASA'), cols.index('MATERIAL')
cols[b], cols[a] = cols[a], cols[b]
x2 = x2[cols]
x2
```

## Identify the customer that has more than 30 transaction

```python
x3 = x2[x2['JASA']+x2['MATERIAL']+x2['MATERIAL & JASA'] + x2['ONGKIR'] >= 30]

x3['total'] = x3['JASA']+x3['MATERIAL']+x3['MATERIAL & JASA'] + x3['ONGKIR']
x3.sort_values(by = 'total', ascending = True, inplace = True)
x3.drop(columns ='total',inplace = True)
x3
```

```python
ax = x3.plot(kind='barh',stacked=True, figsize=(12,6), xticks = np.arange(10, 101, 10).tolist())
plt.title('Rincian lebih dari 30 transaksi yang LUNAS',fontsize= 15)
plt.ylabel('Nama Customer', fontsize= 15)
plt.xlabel('Total Transaksi', fontsize= 15)

for rect in ax.patches:
    # Find where everything is located
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()
    
    # The width of the bar is also not pixels, it's the
    # number of transaction. So we can use it as the label!
    label_text = width
    
    # ax.text(x, y, text)
    label_x = x + width / 2
    label_y = y + height / 2
    ax.text(label_x, label_y, label_text, ha='center', va='center')

plt.show()
```
