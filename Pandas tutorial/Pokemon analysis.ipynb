## Filtering data

```python
new_df = df.loc[(df['Type 1'] == 'Poison') & (df['Type 2']== 'Dark')] # filter type 1 = poison and type 2 = dark
#new_df = new_df.reset_index(drop =True) 
new_df.reset_index(drop =True, inplace = True) # reset the index to 0, 1 ,2 ....
new_df
```

```python
#df.loc[df['Name'].str.contains('Mega')] # filter all name contains 'Mega'
df.loc[~df['Name'].str.contains('Mega')] # filter name not contain Mega just put ~
```

```python
#Use regex to filter 
import re
df.loc[df['Type 1'].str.contains('Fire|Dark', regex=True)] # Filter type 1 'fire or dark' using regex
#can add 'flags=re.I' in str.contains() to ignore capitalisation
```

```python
df.loc[df['Name'].str.contains('Pi[a-z]*', regex=True)] # Filter name contains 'Pi' using regex
```

## Conditional changes

```python
df.loc[df['Total'] > 500, ['Generation','Legendary']]= 'TEST VALUE'
#if Total > 500 then the Legendary and Generation column = TEST VALUE
df
```

## Aggregate Statistics

```python
df.groupby('Type 2').mean().sort_values('Speed',ascending=True)
```

```python
df.groupby('Type 1').sum()
```

```python
df.groupby('Type 1').count()
# we have 69 bugs, 31 dark ....
#however in 'Type 2' column there is missing data cause blank for 'Type 2'
```

```python
#we can clean it by create a new column called 'count'
df['Count']=1
#df.head()
df.groupby('Type 1').count()['Count']
```
