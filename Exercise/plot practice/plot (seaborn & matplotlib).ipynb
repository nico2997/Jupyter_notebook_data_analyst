```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns. load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
```

### try plot tip colummn in histogram

```python
tips.tip.plot(kind='hist')
plt.show()
```

### plot smoker and plot it asbar plot

```python
total_smoker = tips.smoker.value_counts()
total_smoker
```

```python
total_smoker.plot(kind='bar')
```

### Seaborn plot

```python
sns.countplot(x = 'smoker', data = tips)
```

```python
sns.distplot(tips.total_bill, kde =False)

# distrbution plot
```

```python
# linear regression plot
sns.lmplot(x = 'total_bill', y ='tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue='sex')

# hue color is to separate base on the selected name
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue='sex', fit_reg = False)

# fit_reg = False wilol get rid off the line
```

```python
# we can subset it again using col and row function 
# here we want to subset it between smoker and non smoker using col
# and different days using row

sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue='sex', fit_reg = False, col ='smoker')
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue='sex', fit_reg = False, col ='smoker', row = 'day')

# make sure your col and row function are categorical data
```

```python
# we can build a empty grid with desirable components we set

facit = sns.FacetGrid(tips, col ='time', row = 'smoker', hue ='sex')

# we fill it now using the map function (plt.....)
facit.map(plt.scatter, 'total_bill', 'tip')
```

### If we desire to plot multiple distinct plot, we can use fig, ax tuple for matplotlib

```python
fig, ax = plt.subplots(1,1)

# return 1 x 1 figure

# x-axis tip ; y-axis total bill
ax.scatter(tips.tip, tips.total_bill)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)

# return 1 x 2 figure

# x-axis tip ; y-axis total bill
ax1.scatter(tips.tip, tips.total_bill)
ax2.hist(tips.smoker)
```
