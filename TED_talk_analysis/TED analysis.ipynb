```python
def str_to_list(rating_str):
    return ast.literal_eval(rating_str)

str_to_list(ted.ratings[0])
```

```python
ted['rating_list'] = ted.ratings.apply(lambda x : str_to_list(x))
ted.rating_list[0]

# rating_list is a list
```

# 6. Count the total numbers of ratings received by each talk

- new column called num_ratings

```python
for list in ted.rating_list[0]:
    print(list['count'])
```

```python
ted.rating_list[0][0]['count']
```

```python
def get_total_ratings(number_of_ratings):
    num = 0
    for list in number_of_ratings:
        num = num + list['count']
    return num
```

```python
get_total_ratings(ted.rating_list[0])
```

```python
ted['num_ratings'] = ted.rating_list.apply(lambda x: get_total_ratings(x))
```

```python
ted.head()
```

```python
ted.num_ratings.describe()

# just double check the total count we have which is correct 2550
```

# 7. Which occupation delivers the funniest TED talk on average?

```python
ted.drop(columns = 'comments_per_view')
ted.drop(columns = 'views_per_comment')
```

```python
ted.dtypes
```

```python
ted.speaker_occupation
```

```python
ted.ratings.str.contains('Funny').value_counts()
# check if the funny contain in every data
```

```python
def get_funny_ratings(input):
    for list in input:
        if list['name'] == 'Funny':
            return list['count']
```

```python
get_funny_ratings(ted.rating_list[0])
```

```python
ted['funny_ratings']=ted.rating_list.apply(lambda x: get_funny_ratings(x))
```

```python
ted['funny_rates'] = ted['funny_ratings'] / ted['num_ratings']
```

```python
ted.speaker_occupation.value_counts()
```

```python
ted.groupby('speaker_occupation')['funny_rates'].agg(['count', 'mean']).sort_values('count')
```

```python
ted.speaker_occupation.describe()
#missing value for the occupation 
```

## Focus on occupations that well represented data

```python
occupation_counts = ted.speaker_occupation.value_counts()
occupation_counts
```

```python
top_occupations = occupation_counts[occupation_counts >= 5].index
top_occupations

#index can be treated like a list
```

```python
ted_top_occupation = ted[ted.speaker_occupation.isin(top_occupations)]
ted_top_occupation.shape

# we filter with at least 5 person for occupation
```

```python
ted_top_occupation.groupby('speaker_occupation')['funny_rates'].agg(['count', 'mean']).sort_values('mean')
```
