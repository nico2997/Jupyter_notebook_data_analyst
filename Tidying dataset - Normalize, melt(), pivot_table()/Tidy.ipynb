# 2. Create a cases and country column 

```python
ebola = pd.read_csv('country_timeseries.csv')
ebola.head()
```

```python
ebola_long = pd.melt(ebola, id_vars = ['Date', 'Day'])
ebola_long.head()
```

```python
variable_split = ebola_long.variable.str.split('_')
variable_split.head()
```

```python
ebola_long['stats'] = variable_split.str.get(0)
ebola_long['country'] = variable_split.str.get(1)
```

```python
ebola_long
```

```python
variable = ebola_long.variable.str.split('_', expand = True)
variable 
# expand = True to return dataframe instead of list
```

```python
ebola_long[['stats_e', 'country_e']]= (ebola_long.variable
                                       .str
                                       .split('_', expand = True))
ebola_long.head()
```

# 3. get rid of duplicates by changing column name using pivot_table()

```python
weather = pd.read_csv('weather.csv')
weather.head()
```

```python
weather_long = pd.melt(weather, id_vars = ['id', 'year', 'month', 'element'],
              var_name = 'day', value_name ='temp')
weather_long.tail()
```

```python
weather_tidy = weather_long.pivot_table(index = ['id', 'year', 'month', 'day']
                                       , columns = 'element'
                                       , values = 'temp'
                                       ,dropna = True)
weather_tidy
```

```python
# this will return nice and tidy dataframes
weather_tidy.reset_index()
```

# 4. Separate dataset to simpler dataset which has no duplicates (Normalization)

```python
bil_melt.head()
```

```python
bil_melt.loc[bil_melt['track'] == 'Loser']
# this return all same song like duplicates for solid 76 times 
```

```python
billboard_song = bil_melt[['year', 'artist', 'track', 'time']]
billboard_song
```

```python
# we drop duplicates
billboard_song.drop_duplicates(inplace = True)
billboard_song
```

```python
billboard_song['id'] = range(len(billboard_song))
billboard_song.head()
```

```python
billboard_song.to_csv('billboard_song.csv', index = False)
```

```python
billboard_ratings = bil_melt.merge(billboard_song
                                   , on = ['year', 'artist', 'track', 'time']
                                  )
billboard_ratings
```

```python
billboard_ratings = billboard_ratings[['id', 'date.entered', 'week', 'ratings']]
billboard_ratings
```

```python
billboard_ratings.to_csv('billboard_ratings.csv', index = False)
```
