## Random Forest Model 

```python
# We want to predict 'survived' parameter using ["Pclass", "Sex", "SibSp", "Parch"]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
y= train.Survived
y
```

```python
dependent = ["Pclass", "Sex", "SibSp", "Parch"]
```

```python
x = pd.get_dummies(train[dependent])
x
```

```python
x_test = pd.get_dummies(test[dependent])
x_test
```

```python
model = RandomForestClassifier(n_estimators = 100, max_depth = 5, random_state = 1)
```

```python
model.fit(x,y)
```

```python
prediction = model.predict(x_test)
prediction
```

```python
output = pd.DataFrame({'Passenger ID' : test.PassengerId, 'Survived' : prediction })
output
```
